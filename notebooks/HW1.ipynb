```python
#Homework 1 Astr 599
#Part one: loading and exploring the data
import numpy as np
import math
datfile = '../data/asteroids5000.csv'
data = np.genfromtxt(datfile, delimiter=',')
print(data.shape)
```

```python
#How many nans are there? 

binMask = ~np.isnan(data)
numMissing = sum(sum(binMask))
print(numMissing)
```

```python
dataTrans = np.ndarray.transpose(data)
completeCols = ~np.isnan(sum(dataTrans))
print(sum(completeCols))
```

```python
#Create a new array containing only the rows with no missing columns

newdata = data[completeCols]
newdata.shape
```

```python
#compute the max, min, mean, and st dev of values in each column

maxCols = newdata.max(0)
minCols = newdata.min(0)
meanCols = newdata.mean(0)
stdCols = newdata.std(0)
print('max = ', maxCols)
print('min = ', minCols)
print('mean = ', meanCols)
print('standard deviation = ', stdCols)
```
